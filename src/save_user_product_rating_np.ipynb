```python
### Imports
import pandas as pd
import numpy as np
```

```python
base_path="../data/"
```

```python
# Loading Orders_Users Dataset
# TODO : Remove ,nrows=100000 to run for full dataset
fields=["order_id","user_id"]
df_orders_users = pd.read_csv("../data/orders.csv",usecols=fields)
```

```python
# Loading Orders_Products Dataset
# TODO : Remove ,nrows=100000 to run for full dataset
fields=["order_id","product_id"]
df_orders_products = pd.read_csv("../data/order_products__prior.csv",usecols=fields)
```

```python
df_orders_users.head()
```

```python
df_orders_products.head()
```

```python
# user,product Merge
fields=["user_id","product_id"]
df_users_products=df_orders_users.merge(df_orders_products,on='order_id')[fields]
df_users_products.head()
```

```python
# Saving into pickle as user_product_counts
file_name_user_product=base_path+"user_product"
df_users_products.to_pickle(file_name_user_product)
```

```python
# converting into user,product,counts
df_user_product_count=df_users_products.groupby(fields).size().to_frame(name = 'count').reset_index()
df_user_product_count.head()
```

```python
# converting into user_product Utility Matrix
df_user_product_matrix=df_user_product_count.pivot(index='user_id', columns='product_id', values='count').fillna(0)
# .as_matrix()

df_user_product_matrix.head()
```
